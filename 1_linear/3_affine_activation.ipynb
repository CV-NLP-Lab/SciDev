## Краткий экскурс из линейной алгебры

**1. Уравнение прямой**: $ax + by + c = 0$

**2. Однородные координаты (ОК)**

Переход в ОК: $ [x, y] -> [x, y, 1] $

Обратно: $[x, y, z] -> [x/z, y/z]$

Свойство: в ОК $\forall \lambda \ne 0:  [x,y,z] = [\lambda x, \lambda y, \lambda z] $

При $z=0$ точка является точкой на бесконечности. Все точки на бесконечности составляют прямую на бесконечности.

**3. Прямая в ОК**

- Прямая: $l = [a, b, c]$
- Нормаль к прямой $l$: $n = [a, b]$
- Точка в ОК: $p = [x, y, 1]$

Уравнение прямой: $(p, l) = 0$ или $(n, p') = d$

**4. Расстояние от точки до прямой**

$ dist(p, l) = \frac {(p, l)}  {\|n\|} $

**5. Для нейрона:**
- $σ$ - функция активации
- $w$ - веса нейрона (без сдвига)
- $b$ - сдвиг нейрона
- $x$ - входной вектор (без размерности на сдвиг)
- $y$ - выход нейрона

**Действие нейрона:**

$y = σ((w, x) + b) = σ(\|w\| \times dist(w, x) + b) = σ(a \times d + b)$

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

wing = 10
x = np.linspace(-wing, wing, wing * 10 + 1)
```

### Действие множителя

```python
plt.axvline(0,   linestyle=':')
plt.axhline(0.5, linestyle=':')

t = 0.01
for factor in [1, 1/2, 1/4, 1/8, 1/16]:
    y = sigmoid(x * factor)
    line = plt.plot(x, y, label=f'factor=1/{1/factor:.0f}')
    
    p_min = np.argmax(y > t)
    p_max = np.argmax(y > 1-t)
    print(factor, x[p_min], x[p_max])
    if p_min > 0:
        plt.axvline(x[p_min], linestyle='--', color=line[0].get_color())
    if p_max > 0:
        plt.axvline(x[p_max], linestyle='--', color=line[0].get_color())

_ = plt.legend()
```

### Действие слагаемого

```python
plt.axhline(0.5, linestyle=':')

for term in [-5, 0, 5]:
    plt.axvline(-term, linestyle=':')
    plt.plot(x, sigmoid(x + term), label=f'term={term}')
_ = plt.legend()
```
